# Theory

**Van Westendorp's Price Sensitivity Meter**
didefinisikan sebagai teknik bagi market researcher untuk mengukur persepsi konsumen tentang harga produk atau layanan. Penemu pendekatan analisis harga ini adalah seorang ekonom Belanda Peter Van Westendorp. Teknik ini dilakukan dengan metode survei dan mengajukan empat pertanyaan spesifik untuk membantu mengungkap harga terbaik untuk produk tersebut.

**Empat pertanyaan tersebut adalah:**


*   Q1. Pada harga berapa Anda akan menganggap produk tersebut terlalu mahal sehingga Anda tidak akan membelinya? **(Too expensive)**
*   Q2. Pada harga berapa Anda akan menganggap produk tersebut diberi harga terlalu rendah dan karena itu Anda akan merasa bahwa kualitas produk atau layanannya buruk? **(Too cheap)**
*   Q3. Pada harga berapa Anda akan menganggap produk tersebut menjadi terlalu mahal, tetapi Anda harus mempertimbangkan untuk membelinya? **(Expensive)**
*   Q4. Pada harga berapa Anda akan menganggap produk itu sebagai barang murah bagus untuk dibeli? (**Cheap)**

```r
library('pricesensitivitymeter')
```

```r
output_psm <- psm_analysis(toocheap = "toocheap",
                           cheap = "cheap",
                           expensive = "expensive",
                           tooexpensive = "tooexpensive",
                           data = datapricing,
                           validate = TRUE)
```

```r
summary(output_psm)
```

```r
psm_plot(output_psm, "myplot")
```
